```python
# Importing relevant libraries
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch
import pandas as pd
import matplotlib.font_manager as font_manager
```

```python
# Importing the CSV file 
file_path = '/Users/aaronbopara/Downloads/data/portfolio_work/python_football_data_/afcshots_cfc_2324.csv'
shots_df = pd.read_csv(file_path)
shots_df
```

```python
shots_df['player_x'] = 100 - shots_df['player_x']
shots_df
```

```python
total_shots = shots_df.shape[0]
total_goals = shots_df[shots_df['shotType'] == 'goal'].shape[0]
total_xg = shots_df['xg'].sum()
xg_per_shot = total_xg / total_shots
```

```python
# Import fonts to use, checking branding guides for the football club you are making a graphic for 
font_path = '/Users/aaronbopara/Library/Fonts/ufonts.com_agencyfb-black.ttf'
font_props = font_manager.FontProperties(fname = font_path )
```

```python
# Initialise our pitch with the desired settings 
pitch = VerticalPitch(pitch_type='opta', pitch_color='#f1c58c', line_color='#1d1b20', half=True, pad_bottom=0.2, corner_arcs=True, linewidth=2, line_alpha=1)

# Create the figure and axes
fig = plt.figure(figsize=(8, 7))  # Adjusted figure size to reduce bottom space
fig.patch.set_facecolor('#f1c58c')

# Title and description area
ax1 = fig.add_axes([0, 0.75, 1, 0.23])  # Adjust height and position to remove bottom space
ax1.set_facecolor('#f1c58c')
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)

ax1.text(
    x=0.5, 
    y=0.77, 
    s='Arsenal Vs Chelsea 2023/2024', 
    fontsize=30, 
    fontproperties=font_props, 
    fontweight='bold', 
    color='#1b191e', 
    ha='center'
)
ax1.text(
    x=0.5, 
    y=0.55, 
    s='Arsenal Shot Map', 
    fontsize=20,
    fontweight='bold',
    fontproperties=font_props, 
    color='#1b191e', 
    ha='center'
)

# Quality chance scatter points
ax1.text(x=0.25, y=0.35, s='Low Quality Chance', fontsize=15, fontproperties=font_props, color='#1d1b20', ha='center')
ax1.scatter(x=0.37, y=0.4, s=100, color='#269584', edgecolor='#1d1b20', linewidth=1.5)
ax1.scatter(x=0.42, y=0.4, s=200, color='#269584', edgecolor='#1d1b20', linewidth=1.5)
ax1.scatter(x=0.48, y=0.4, s=300, color='#269584', edgecolor='#1d1b20', linewidth=1.5)
ax1.scatter(x=0.54, y=0.4, s=400, color='#269584', edgecolor='#1d1b20', linewidth=1.5)
ax1.scatter(x=0.6, y=0.4, s=500, color='#269584', edgecolor='#1d1b20', linewidth=1.5)
ax1.text(x=0.75, y=0.35, s='High Quality Chance', fontsize=15, fontproperties=font_props, color='#1d1b20', ha='center')

# Adjusted positions to align the elements horizontally
ax1.text(x=0.21, y=0.1, s='Goal', fontsize=15, fontproperties=font_props, color='#1d1b20', ha='right')
ax1.scatter(x=0.23, y=0.15, s=100, color='#269584', edgecolor='#1d1b20', linewidth=1.5, alpha=.7)

ax1.text(x=0.41, y=0.1, s='Missed Shot', fontsize=15, fontproperties=font_props, color='#1d1b20', ha='right')
ax1.scatter(x=0.43, y=0.15, s=100, color='#ffffff', edgecolor='#1d1b20', linewidth=1.5)

ax1.text(x=0.6, y=0.1, s='Saved Shot', fontsize=15, fontproperties=font_props, color='#1d1b20', ha='right')
ax1.scatter(x=0.62, y=0.15, s=100, color='#9f111f', edgecolor='#1d1b20', linewidth=1.5, alpha=.7)

ax1.text(x=0.8, y=0.1, s='Blocked Shot', fontsize=15, fontproperties=font_props, color='#1d1b20', ha='right')
ax1.scatter(x=0.82, y=0.15, s=100, color='#1d1b20', edgecolor='#1d1b20', linewidth=1.5)

# Hide the axis
ax1.set_axis_off()

# Shot map (Pitch)
ax2 = fig.add_axes([0, 0.05, 1, 0.7])  # Adjusted position to fill bottom space more efficiently
ax2.set_facecolor('#f1c58c')
pitch.draw(ax=ax2)

# Plotting shot circles 
for index, shot in shots_df.iterrows():
    if shot['shotType'] == 'goal':
        shot_color = '#269584'
    elif shot['shotType'] == 'save':
        shot_color = '#9f111f'
    elif shot['shotType'] == 'miss':
        shot_color = '#ffffff'
    elif shot['shotType'] == 'block':
        shot_color = '#1b191e'
    
    shot_size = shot['xg'] * 2000  # Scaling xG for visibility 
    ax2.scatter(shot['player_y'], shot['player_x'], s=shot_size, color=shot_color, edgecolors='#1b191e', linewidth=1.5)

# Stats area
ax3 = fig.add_axes([0, -0.15, 1, 0.3])  
ax3.set_facecolor('#f1c58c')
ax3.set_xlim(0, 1)
ax3.set_ylim(0, 1)

ax3.text(x=0.225, y=0.43, s='Shots', fontsize=20, fontproperties=font_props, fontweight='bold', color='#1d1b20', ha='left')
ax3.text(x=0.25, y=0.25, s=f'{total_shots}', fontsize=16, fontproperties=font_props, color='#269584', ha='left')
ax3.text(x=0.385, y=0.43, s='Goals', fontsize=20, fontproperties=font_props, fontweight='bold', color='#1d1b20', ha='left')
ax3.text(x=0.410, y=0.25, s=f'{total_goals}', fontsize=16, fontproperties=font_props, color='#269584', ha='left')
ax3.text(x=0.55, y=0.43, s='xG', fontsize=20, fontproperties=font_props, fontweight='bold', color='#1d1b20', ha='left')
ax3.text(x=0.5450, y=0.25, s=f'{total_xg:.2f}', fontsize=16, fontproperties=font_props, color='#269584', ha='left')
ax3.text(x=0.675, y=0.43, s='xG/Shot', fontsize=20, fontproperties=font_props, fontweight='bold', color='#1d1b20', ha='left')
ax3.text(x=0.710, y=0.25, s=f'{xg_per_shot:.2f}', fontsize=16, fontproperties=font_props, color='#269584', ha='left')
ax3.set_axis_off()

# Export the figure and display 
plt.savefig('afc_shotmap_cfc.png', dpi=300, bbox_inches='tight')
plt.show()
```
